```python
import pandas as pd
import numpy as np
```

# Para que este código funcione:
- A pasta `EA-MD-QD-2026-02` deve estar no diretório `dados`
- Este ficheiro deve estar no mesmo nível do diretório `dados`
- Devem rodar duas vezes o ficheiro `routine_data.py` para salvarem a base de dados mensal e trismestral
- Criar o diretório `data_TR3` dentro do diretório `EA-MD-QD-2026-02`
- E, voilá!

```python
dir = "dados/EA-MD-QD-2026-02/data_TR2"

dfM = pd.read_excel(f"{dir}/PTdataM_NA__TR2.xlsx")
dfQ = pd.read_excel(f"{dir}/PTdataQ_NA__TR2.xlsx")

dfA = pd.read_excel("dados/EA-MD-QD-2026-02/raw/PTdata.xlsx")
```

```python
dfM.head()
```

```python
dfA.head()
```

```python
dfMonthly = dfA[['Time',] + dfM.drop('Date', axis=1).columns.tolist()].copy()
dfMonthly.head()
```

```python
dfQuarter = dfA[['Time',] + dfQ.drop('Date', axis=1).columns.tolist()].copy()
dfQuarter.head()
```

```python
dfMonthly.to_excel("dados/EA-MD-QD-2026-02/data_TR3/monthly-PT.xlsx", index=False)
dfQuarter.to_excel("dados/EA-MD-QD-2026-02/data_TR3/quarter-PT.xlsx", index=False)
```
