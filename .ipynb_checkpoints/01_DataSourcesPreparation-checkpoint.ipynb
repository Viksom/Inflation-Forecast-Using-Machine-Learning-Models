## **2. EA-MD-QD**

> **Nota:** Esta secção serve exclusivamente para extrair os dados em **nível (não estacionarizados)** da EA-MD-QD, separados por frequência (mensal e trimestral), para utilização no `DataIntegration.ipynb`. Os ficheiros TR2 (`data_TR2/`) — que contêm dados já transformados/estacionarizados pelo `routine_data.py` — são usados apenas como **lista de colunas** para identificar quais variáveis são mensais e quais são trimestrais; os seus valores nunca entram no dataset final.

## **3. PPI Portugal**

- **Dados brutos**: [Producer Price Index PT — EconDB](https://www.econdb.com/series/PPIPT~TOYA/portugal-producer-price-index/?from=1999-10-11&to=2026-04-20) - série mensal do Índice de Preços no Produtor para Portugal
- **Imputação manual (Jan–Dez 2000)**: [CountryEconomy — Portugal PPI 2000](https://countryeconomy.com/prices/producer/portugal?year=2000) - os primeiros 12 meses não têm ano anterior para calcular a variação homóloga, pelo que foram imputados manualmente a partir desta fonte que contém dados da variação homóloga desde 1990 (não se utilizou apenas esta fonte dado que para extrair os valores tem de ser de forma manual)

```python
df_PPI = pd.read_csv("dados/Portugal_PPI_mensal_2000_2026.csv")

# Calcular a taxa de variação homóloga
df_PPI["Date"] = pd.to_datetime(df_PPI["Date"])
df_PPI = df_PPI.sort_values("Date").reset_index(drop=True)

df_PPI["PPIPT_tx_var_homologa"] = (
    (df_PPI["PPIPT"] - df_PPI["PPIPT"].shift(12))
    / df_PPI["PPIPT"].shift(12)
) * 100
df_PPI["PPIPT_tx_var_homologa"] = df_PPI["PPIPT_tx_var_homologa"].round(1)

# Os 12 primeiros meses (Jan-Dez 2000) não têm ano anterior — imputação manual
valores_imputados = [21.5, 21.9, 21.2, 20.0, 17.9, 17.2, 17.0, 14.6, 12.8, 13.1, 12.1, 10.3]
df_PPI.loc[:11, "PPIPT_tx_var_homologa"] = valores_imputados

df_PPI["PPIPT"] = df_PPI["PPIPT_tx_var_homologa"]
df_PPI = df_PPI.drop(columns=["PPIPT_tx_var_homologa"])

df_PPI.head(15)
```

```python
plt.figure(figsize=(12, 5))
plt.plot(df_PPI["Date"], df_PPI["PPIPT"], linewidth=1.8)
plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.title("Producer Price Index PT — Taxa de Variação Homóloga (%)")
plt.xlabel("Data")
plt.ylabel("Taxa de variação homóloga (%)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
df_PPI.to_csv("dados/PPI_PT.csv", index=False)
```
