### Como estão as datas definidas?
- **target**: `Período de referência` - DD/MM/AAAA
- **fred-md**: `sasdate` - M/D/AAAA
- **epu**: `Unnamed: 0` - AAAAmM
- **unemployment**: `Time period` - AAAA-month_short_name
- **ea-md & ea-qd**: `Time` - AAAA-MM-DD
- **ppi**: `Date` - AAAA-MM-DD

---
### Principais problemas das Bases de Dados
- **target:** sem problemas, em princípio, só precisamos de duas variáveis;
- **fred-md:** eliminar a linha 0 e só precisamos de 8 variáveis da base de dados;
- **epu:** eliminar as duas últimas colunas e mudar o nome da primeira para Date ou algo assim;
- **unemployment:** eliminar a última coluna;
- **ea-md & ea-qd:** sem stress para a primeira, a segunda vai precisar que apliquemos interpolação.

### Coisas importantes a verificar
- Para cada uma das bases de dados os valores correspondem ao que foi calculado no último dia do mês ou no primeiro do mês seguinte?
    - **target:** Último dia do mês (ponderando colocar no primeiro dia do mês corrente, por convenção)
    - **fred-md, ppi, ea-md & ea-qd:** Primeiro dia do mês
    - **epu & unemployment:** Desconhecido, é importante que se verifique e confirme a informação (acredito que seja o último, mas por convenção deverá ficar mesmo o primeiro dia)
