## **3. Pré-processamento**

### Formatos de Data por Fonte

- **target**: `Período de referência` — DD/MM/AAAA
- **fred-md**: `sasdate` — M/D/AAAA
- **epu**: `Unnamed: 0` — AAAAmM
- **unemployment**: `Time period` — AAAA-Mês
- **ea-md & ea-qd**: `Time` — AAAA-MM-DD
- **exchange**: `Período de referência` — AAAA-MM-DD
- **ppi**: `Date` — AAAA-MM-DD

Todas as datas são normalizadas para o primeiro dia do mês (`AAAA-MM-01`).

### **3.1 Seleção de Colunas**

Da FRED-MD são retidas apenas 15 colunas relevantes (produção industrial, mercado de trabalho, taxas de juro, preços e mercados financeiros) identificadas nas fontes de estudo. Das restantes fontes eliminam-se colunas auxiliares de metadados — ficam apenas a série temporal e a coluna de data.

```python
df_target = df_target[['Período de referência', 'Valor']]

df_exchange = df_exchange[['Período de referência', 'Valor']]

cols_fred_global = [
    'sasdate', 'INDPRO', 'IPFINAL', 'IPCONGD',
    'AWOTMAN', 'AWHMAN', 'FEDFUNDS', 'GS1', 'GS5',
    'GS10', 'AAA', 'BAA', 'OILPRICEx', 'CUSR0000SAC',
    'PCEPI', 'S&P 500'
]
df_fred_md = df_fred_md.iloc[1:].reset_index(drop=True)[cols_fred_global]

df_epu.drop(df_epu.columns[-2:], axis=1, inplace=True)

df_unemployment.drop(df_unemployment.columns[-1], axis=1, inplace=True)

# df_ea_md # Está good
# df_ea_qd # Irá ser feita Interpolação e Auto-Arima
```

### **3.2 Normalização das Datas**

Todas as colunas de data são renomeadas para `Date`, os formatos heterogéneos são convertidos para `datetime` e os valores normalizados para o primeiro dia de cada mês — convenção adoptada em todo o projeto.

```python
default_date = 'Date'

df_target.rename({'Período de referência': default_date}, axis=1, inplace=True)
df_fred_md.rename({'sasdate': default_date}, axis=1, inplace=True)
df_epu.rename({'Unnamed: 0': default_date}, axis=1, inplace=True)
df_unemployment.rename({'Time period': default_date}, axis=1, inplace=True)

df_ea_md.rename({'Time': default_date}, axis=1, inplace=True)
df_ea_qd.rename({'Time': default_date}, axis=1, inplace=True)

df_exchange.rename({'Período de referência': default_date}, axis=1, inplace=True)
```

```python
df_unemployment.dropna(axis=0, inplace=True)
df_unemployment.reset_index(drop=True, inplace=True)
```

```python
df_target['Date'] = pd.to_datetime(
    df_target['Date'],
    dayfirst=True
)

df_fred_md['Date'] = pd.to_datetime(
    df_fred_md['Date']
)

df_ea_md['Date'] = pd.to_datetime(
    df_ea_md['Date']
)

df_ea_qd['Date'] = pd.to_datetime(
    df_ea_qd['Date']
)

df_ppi['Date'] = pd.to_datetime(
    df_ppi['Date']
)

df_epu['Date'] = pd.to_datetime(
    df_epu['Date'].str.replace('m', '/'),
    format='%Y/%m'
)

df_unemployment['Date'] = pd.to_datetime(
    df_unemployment['Date'],
    format='%Y-%b'
)
```

```python
df_exchange['Date'] = pd.to_datetime(
    df_exchange['Date'],
    format='%Y-%m-%d'
)
```

```python
df_target['Date'] = df_target.Date.dt.to_period("M").dt.to_timestamp()

df_exchange['Date'] = df_exchange.Date.dt.to_period("M").dt.to_timestamp()
```

```python
dfs = {
    'target': df_target, 'fred-md': df_fred_md,
    'ea-md': df_ea_md, 'ea-qd': df_ea_qd,
    'epu': df_epu, 'unemployment': df_unemployment,
    'exchange': df_exchange, 'ppi': df_ppi
}

for df_name, df in dfs.items():
    print(f"Data mínima de {df_name} é `{df['Date'].min()}`")
```

```python
for df_name, df in dfs.items():
    print(f"Data máxima de {df_name} é `{df['Date'].max()}`")
```

```python
total = 0
for df_name, df in dfs.items():
    print(f"{df_name} dataset: {df.shape[1] - 1} coluna/s")
    total += df.shape[1] - 1 # Eliminar data

print("=" * 30)
print(f"Total: {total} colunas")
```

## **4. Integração dos Dados**

As colunas de cada fonte são renomeadas com o sufixo da origem (e.g. `INDPRO_fred-md`) para evitar colisões. O merge é feito por left join sobre a EA-MD-QD mensal. O dataset resultante é truncado a Out 2025, data da última observação para o indice de incerteza politica.

```python
df_target.rename({'Valor': 'inflation'}, axis=1, inplace=True)
df_unemployment.rename({'Value': 'unemployment_up_15'}, axis=1, inplace=True)
df_exchange.rename({'Valor': 'exchange'}, axis=1, inplace=True)
```

```python
for df_name, df in dfs.items():
    rename = dict()
    for col in df.columns[1:]:
        rename[col] = f'{col}_{df_name}'
    df.rename(rename, axis=1, inplace=True)
```

```python
df = pd.merge(df_ea_md, df_ea_qd, 'left', 'Date')
df = pd.merge(df, df_fred_md, 'left', 'Date')
df = pd.merge(df, df_epu, 'left', 'Date')
df = pd.merge(df, df_unemployment, 'left', 'Date')
df = pd.merge(df, df_exchange, 'left', 'Date')
df = pd.merge(df, df_ppi, 'left', 'Date')
df = pd.merge(df, df_target, 'left', 'Date')
df.head()
```

```python
df = df.loc[df.Date <= '2025-10-01']
```

## **5. Engenharia de Atributos**

O preço do petróleo da FRED-MD está em USD. Para torná-lo comparável com a economia da Área do Euro, converte-se para euros dividindo pela taxa de câmbio EUR/USD do mesmo mês.

```python
df['OILPRICEx_fred-md_Eur'] = df['OILPRICEx_fred-md'] / df['exchange_exchange']
```
